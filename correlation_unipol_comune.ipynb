```python
import numpy as np
from bs4 import BeautifulSoup, Comment
import datetime
from datetime import datetime
from pymongo import MongoClient
import matplotlib.pyplot as plt
from datetime import timedelta
%matplotlib inline
```

```python
import xml.etree.ElementTree as ET
```

```python
tree = ET.parse('/home/amassaro/Downloads/xml_incidenti_gen-dic_2014.xml')
root=tree.getroot()
child=list(root)[1]
```

```python
k=0
t=0
keys=[]
for j in child:
    t+=1
    keys+=j.attrib.keys()
print list(set(keys))        
```

```python
time=[]
for j in child:
    if 'DataOraIncidente' in j.attrib.keys():
        time.append(j.attrib['DataOraIncidente'])

time=np.array([t[:10]+' '+t[11:] for t in time])
time=np.array([datetime.strptime(t,'%Y-%m-%d %H:%M:%S') for t in time])
day_time_comune=np.array([[t.weekday()+1,t.hour] for t in time])
```

```python
day_time_unipol=[]
f=open('/home/amassaro/Downloads/BDC2015_UnipolsaiClaims2014_RM.csv')
k=0
for line in f:
    if k>0:
        t=[]
        line=line.split(',')
        t.append(int(line[2]))
        t.append(int(line[3][:-2]))
        day_time_unipol.append(list(t))
    k+=1
day_time_unipol=np.array(day_time_unipol)
```

```python
print('hourly probability of car-crash for 2014')
plt.hist(day_time_unipol[:,1],bins=list(range(25)),normed=True,alpha=0.3,color='yellow',label='unipol')
plt.ylim([0,.1])
plt.legend(loc='upper right')
plt.gcf().set_size_inches(18, 5)
plt.show()

plt.hist(day_time_comune[:,1],normed=True,bins=list(range(25)),alpha=0.5,color='black',label='comune')
plt.legend(loc='upper right')
plt.ylim([0,.1])
plt.gcf().set_size_inches(18, 5)
plt.show()

H_unipol=np.histogram(day_time_unipol[:,1],bins=list(range(25)),normed=True)[0]
H_comune=np.histogram(day_time_comune[:,1],bins=list(range(25)),normed=True)[0]
plt.hist(day_time_unipol[:,1],bins=list(range(25)),normed=True,alpha=0.3,color='yellow',label='unipol')
plt.hist(day_time_comune[:,1],normed=True,bins=list(range(25)),alpha=0.5,color='black',label='comune')
plt.ylim([0,.1])
plt.legend(loc='upper right')
plt.gcf().set_size_inches(18, 5)
plt.show()

H_unipol=np.histogram(day_time_unipol[:,1],bins=list(range(25)),normed=True)[0]
H_comune=np.histogram(day_time_comune[:,1],bins=list(range(25)),normed=True)[0]
epsilons=np.abs(H_unipol-H_comune)
print 'max_diff: ',round(max(epsilons),4),'\nmin_diff: ',round(min(epsilons),4),'\navg_diff: ',round(np.average(epsilons),4) 
```
